```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from mpl_toolkits import mplot3d
import seaborn as sns
```

**Load dataset of normal empirical observations**

```python
# Load normal observations dataset

normal = pd.read_csv("Whitby_river_rain_sea_92_17.csv",index_col=[0])
normal.head()
```

**Load dataset of empirical CDF values**

```python
# Load cdf values 

cdf_df = pd.read_csv("Whitby_river_rain_sea_92_17_cdf.csv",index_col=[0])
cdf_df.head()
```

**Pair-wise scatter plots on the off-diagonal of the empirical observations**

**Marginal distributions on the diagonal** 

```python
plt.figure(figsize=(5,6))
sns.pairplot(normal)
plt.show()
```

**Scatter plot of a sample of the CDF values followed by a 3d scatter plot of all variables**

```python
# scatter plots
m = 10

fig, ax = plt.subplots(1,3,figsize=(15,5))
ax[0].scatter(cdf_df['rain'].values[::m],cdf_df['elevation'].values[::m])
ax[0].set_title("Rain vs Sea level empirical CDF values")
ax[0].set_xlabel("Rain")
ax[0].set_ylabel("Sea level")

ax[1].scatter(cdf_df['rain'].values[::m],cdf_df['river'].values[::m])
ax[1].set_title("Rain vs River discharge empirical CDF values")
ax[1].set_xlabel("Rain")
ax[1].set_ylabel("River discharge")

ax[2].scatter(cdf_df['river'].values[::m],cdf_df['elevation'].values[::m])
ax[2].set_title("River vs Sea level empirical CDF values")
ax[2].set_xlabel("River discharge")
ax[2].set_ylabel("Sea level")
plt.show()

fig = plt.figure(figsize = (8, 6))
ax = plt.axes(projection ="3d")
 
# Create Plot
ax.scatter3D(cdf_df['rain'].values[::m],cdf_df['elevation'].values[::m],cdf_df['river'].values[::m])
ax.set_xlabel('river')
ax.set_ylabel('sea level')
ax.set_zlabel('rain')

plt.show()
```

**Rainfall vs river discharge**
The above plot suggest clear tail-dependence between the rainfall and river discharge values. This is explainable by a simple causal process: Rainfall -> runoff -> river level. 

**Rainfall vs Sea level**
The above plot suggests there could be some tail dependence between extreme rainfall and extreme sea levels. A visual inspection hints at more clustering at the upper tails compared with the lower tails yet the relationship still appears weak.

**River vs Sea level**
This plot exhibits no clear patterns to speak of. Would suggest that river and sea levels in the Whitby region were independent of each other over the period 1992-2017.
